### Question 1
Take your dataset from the mid-project (the final dataset with the most features, including your engineered features). Train a logistic regression ML model on your data to predict whether the home or visitor team will win.   


```
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
```

```
# Import the csv file
sports = pd.read_csv('DataSet_With.csv')
sports
```

```
# Explore dataset
sports.value_counts()
```

```
# Quzntize the dataset
sports = pd.get_dummies(sports)
sports
```

```
# Full feature set logistic regression
X = sports.drop('HomeWin', axis = 1)
y = sports['HomeWin']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/2)
X_train
```

```
full_log_reg = LogisticRegression()
full_log_reg.fit(X_train,y_train)
```

```
print('Training accuracy: ', full_log_reg.score(X_train,y_train))
print('Testing accuracy: ', full_log_reg.score(X_test,y_test))
```

```
pd.Series(data=full_log_reg.coef_[0], index=X_train.columns).sort_values(ascending=False)
```
